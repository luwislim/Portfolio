```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score, mean_squared_error, mean_absolute_error, r2_score
from sklearn.preprocessing import PolynomialFeatures
```

```python
#load file 
df = pd.read_csv('responses.csv')
df
```

```python
#mengubah string menjadi angka di Smoking  
df["Smoking"] = df["Smoking"].replace('current smoker', 1)  
df["Smoking"] = df["Smoking"].replace('former smoker', 1) 
df["Smoking"] = df["Smoking"].replace('tried smoking', 0)   
df["Smoking"] = df["Smoking"].replace('never smoked', 0)
#mengubah string menjadi angka di Alcohol
df["Alcohol"] = df["Alcohol"].replace('drink a lot', 2)  
df["Alcohol"] = df["Alcohol"].replace('social drinker', 1)    
df["Alcohol"] = df["Alcohol"].replace('never', 0)
```

```python
#drop data
newDf = df.dropna()
print(newDf.isna().values.any())
```

```python
#declare dataset feature and target 
feature = newDf[["Music","Slow songs or fast songs","Dance","Folk","Country","Classical music","Musical","Pop","Rock","Metal or Hardrock","Punk","Hiphop, Rap","Reggae, Ska","Swing, Jazz","Rock n roll","Alternative","Latino","Techno, Trance","Opera"]]
target = newDf[["Smoking"]]
x, y = np.array(feature), np.array(target)
```

```python
#mengapply PolynominalFeatures untuk selanjutnya dimasukkan ke linear regression
transformer = PolynomialFeatures(degree=2, include_bias=False)
```

```python
#mengfit data
transformer.fit(x)
```

```python
#mengtransform input array 
x= transformer.transform(x)
```

```python
print(x)
```

```python
#mengapply linear regression dan fit
model = LinearRegression().fit(x, y)
```

```python
r_sq = model.score(x, y)
print('coefficient of determination:', r_sq)# hasil R-Squared (bila di bawah 0.5 maka feature tidak berpengaruh kepada target)
print('intercept:', model.intercept_)# hasil coefficient pada b0 (bila di atas 0.05 maka feature tidak berpengaruh kepada target)
print('slope:', model.coef_)# hasil coefficient pada bias setelah b0 yakni b1 b2 b3(bila di atas 0.05 maka feature tidak berpengaruh kepada target)
```

```python
y_pred = model.predict(x)
print('predicted response:', y_pred, sep='\n')
#memprediksi nilai nilai untuk target target
```

```python
print("Mean Absolute Error (MAE): ", mean_absolute_error(y, y_pred))
print("Mean Squared Error (MSE): ", mean_squared_error(y, y_pred))
print("R2: ", r2_score(y, y_pred))
```
